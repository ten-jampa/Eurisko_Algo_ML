# Chapter 4: Simulating Coin Flips

To estimate the probabilities of events by simulating a large number of random experiments and measuring the proportion of trials in which the desired outcome occured. This technique is known as Monte Carlo Simulation. 

```python
# Exercise 
import random 

def coin_flip():
    toss = random.random()
    if toss < 0.5:
        return 'H'
    else:
        return 'T'
        
def multiple_toss(num_flips):
    heads = 0
    tails = 0
    for _ in range(num_flips):
        toss = coin_flip()
        if toss == 'H':
            heads += 1
        else:
            tails += 1
    return heads, tails

def ntrials(num_flips, ntrials = 1000):
    record = []
    for _ in range(ntrials):
        heads, tails = multiple_toss(num_flips)
        record.append(heads)
    return record
    
def sim_probability(num_heads, num_flips, ntrial):
    count = 0
    experiment = ntrials(num_flips, ntrial) #returns the list with number of heads
    for heads in experiment:
        if heads == num_heads:
            count += 1
        else:
            continue
    return count/ntrial

import matplotlib.pyplot as plt

def multiple_expt(num_flips, ntrials):
    num_heads = range(10, num_flips,5)
    probs = []
    for head in num_heads:
        prob = sim_probability(head, ntrials)
        probs.append(probs)
    plt.plot(num_heads, probs)

# multiple_expt(100, 1000)
```

```python
import numpy as np
num_heads = np.arange(1,100)
probs = []
for head in num_heads:
    prob = sim_probability(head,100,2500)
    probs.append(prob)

#print(probs)
print(sum(probs))    
plt.plot(num_heads, probs)
```

```python
# Analytic solution
# Binomial solution 

import scipy.stats
from scipy.stats import binom
probs = []
for head in num_heads:
    prob = binom.pmf(head, 100, 0.5)
    probs.append(prob)
plt.plot(num_heads, probs)
print(sum(probs))
```
